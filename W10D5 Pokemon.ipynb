**È verosimile che la prima colonna dovrebbe essere un indice?** Si, possiamo scegliere di utilizzare la prima colonna con df.set_index[#]

**Confrontiamolo con l'indice messo automaticamente da Pandas: combaciano?** No, il database parte dalla riga 1, Pandas fa partire l'indice dalla posizione 0

**Se no, settare la prima colonna come indice.**

```python
df = df.set_index("#")
```

```python
Legendary = df[df["Legendary"]] #Quali sono i Pokémon leggendari?
print(Legendary)
```

```python
Grass = df.loc[(df["Legendary"] == True) & (df["Type 1"] == "Grass")] #E quali sono i leggendari di tipo 1 Grass?
Grass
```

```python
IceFire = df.loc[(df["Legendary"] == True) & ((df["Type 1"] == "Ice")|(df["Type 1"] == "Fire")) ] 
IceFire  # E leggendari di tipo 1 Ice o Fire?
```

```python
# Trasformiamo la colonna Name nell'indice
df = df.set_index("Name")
df
```

```python
# Quali sono i Pokémon della prima generazione con attacco > 50 e HP < 60?
Gen1 = df.loc[(df["Generation"] == 1) & (df["Attack"] > 50) & (df["HP"] < 60)]
Gen1
```
